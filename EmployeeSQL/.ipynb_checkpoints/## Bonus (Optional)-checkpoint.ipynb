```python
# Import dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sqlalchemy import create_engine
```

```python
# Create engine and connection
engine = create_engine('postgresql://postgres:sana@localhost:5432/employee_sql')
connection = engine.connect()
```

```python
# Create a histogram to visualize the most common salary ranges among employees

# Read .sql file
salaries_df = pd.read_sql('SELECT e.emp_no, s.salary FROM employees as e JOIN salaries as s on e.emp_no = s.emp_no;',
                          connection)

# Determine conditions to plot
salaries = salaries_df.loc[salaries_df['salary']<=90000]

# Plot histogram and set bins
plt.hist(salaries['salary'], bins=5)

# Determine axis and title
plt.xlabel('Salaries')
plt.ylabel('Employees')
plt.title('Employee Salary Ranges - Most Common')
plt.show()
```

```python
# Create a bar chart of average salary by title

# Read .sql file
salaries = pd.read_sql("SELECT * FROM salaries", connection)

# Read .sql file
employees = pd.read_sql("SELECT * FROM employees", connection)

# Merge datasets
combined_data = pd.merge(salaries, employees, on="emp_no", how="inner")

# Group by titles
grouped_by_titles = combined_data.groupby("emp_title_id").mean()

# Plot bar graph
grouped_by_titles.plot(kind='bar',color='lightblue',width=0.9,legend=None)

# Determine axis and title
plt.title("Average Salary By Title")
plt.xlabel("Title ID")
plt.ylabel("Average Salary")
plt.show()
```
